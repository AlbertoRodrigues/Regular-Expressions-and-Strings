## Para capturar uma parte da expressão regular devemos usar () na parte de interesse

```python
import re
```

```python
string = "Clary has 2 friend who she spends a lot time with. Susan has 3 brothers while John has 4 sisters"
```

```python
# Expressão regular de interesse
re.findall(r'[A-Za-z]+\s\w+\s\d+\s\w+', string)
```

```python
# Retornando somente a primeira palavra de interesse
re.findall(r'([A-Za-z]+)\s\w+\s\d+\s\w+', string)
```

### Podemos pegar mais partes complicadas usando mais parentêses, como no exemplo:


```python
re.findall(r'([A-Za-z]+)\s\w+\s(\d+)\s(\w+)', string)
```

## Se quisermos construir uma expressão regular um pouco mais complicada usando o "|" podemos fazer isso usando os parentêses

```python
my_string = "I want to have a pet. But I don't know if I want 2 cats, 1 dog or a bird."
```

```python
# PEgando as palavras cat, dog ou bird, se tiver um número antes dessa palavra
re.findall(r"\d+\s(cat|dog|bird)", my_string)
```

```python
# Pegando o número também
re.findall(r"(\d)+\s(cat|dog|bird)", my_string)
```
